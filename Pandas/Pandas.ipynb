## Loading data into Pandas

```python
import pandas as pd
```

```python
# reading CSV file using pd.read_csv('')
data = pd.read_csv('pokemon_data.csv')

# reading Excel file using pd.read_excel('')
# !pip install openpyxl
# data = pd.read_excel('pokemon_data.xlsx')

# reading txt file using pd.read_csv('') and specify delimeter
# data = pd.read_csv('pokemon_data.txt', delimiter='\t')

print(data)
```

```python
#reading the head(from top) of our data 
#initially threr will be 5 rows but we can specify nuumber
print(data.head(6))
```

```python
#reading the tail(from bottom ) of our data
#initially threr will be 5 rows but we can specify nuumber
print(data.tail(10))
```

## Reading Data in Pandas

```python
# Read Headers
#will print all column names
# print(data.columns) 

# Read specific colummn
print(data[['#', 'Name']][0:10])
```

```python
# Read specific rows
# print(data.iloc[0:5])

#reading rows by iterating through them
# for index, row in data.iterrows():
#     print(index, row['Name'])

# Reading specific location
# print(data.iloc[4, 2]) #iloc is number based
print(data.loc[data['HP'] < 10]) #more in filtering and specific rows
```

## Sorting/ Describing Data

```python
# Specifies the min, max, mean etc
data.describe()
```

```python
# Sorting through data 
# Note: specify ascending=False if want to sort in descending
data.sort_values(by=['Type 1','Attack'], ascending=[1,1])
```

## Making changes to our data

```python
data.head(5)
```
