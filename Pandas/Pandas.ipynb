## Sorting/ Describing Data

```python
# Specifies the min, max, mean etc
data.describe()
```

```python
# Sorting through data 
# Note: specify ascending=False if want to sort in descending
data.sort_values(by=['Type 1','Attack'], ascending=[1,1])
```

## Making changes to our data

```python
data.head(5)
```

```python
#Adding a new column total = HP+ Attack+ Defense+ Sp.Atk+ Sp.Def+ Speed
# data['Total'] = data['HP']+ data['Attack']+ data['Defense']+ data['Sp. Def']+ data['Sp. Atk']+ data['Speed']
data['Total'] = data.iloc[:,4:10].sum(axis=1)
# Above statement added a new column 'total'
print(data[['Name','Total']])
```

```python
data.head()
```

```python
# Dropping/deleting a column_
data = data.drop(columns=['Legendary'])
data.head()
```

```python
# Moving position of columns
columns = list(data.columns.values)
print(columns)
data = data[columns[0:4] + [columns[-1]] + columns[4:11]]
# print(data.head())
```

## Saving our data(Exporting it to desired extension)

```python
# print(data)
data.to_csv('Updated_data.csv', index=False)
# data.to_excel('Updated_data.xlsx', index=False)
# data.to_csv('Updated_data.txt', index=False, sep='\t')
```

## Filtering Data

```python
data.head()
```

```python
# in pandas use '&' for 'and' and '|' for 'or'
# Filtering data 
new_data = data.loc[(data['Type 1'] == 'Grass') & (data['Type 2'] == 'Poison') & (data['Total'] > 500)]
# print(new_data)
# we can export this filtered data eailt using new_data.to_csv('filtered.csv')
# but we notice that the index in file is the old one.So,
new_data = new_data.reset_index(drop=True)
# print(new_data)
new_data.to_csv('filtered.csv')
```

```python
# Editing names by filtering them
#for eg. here we are going to get entries with 'mega' in name.
mega = data.loc[data['Name'].str.contains('Mega')]
print(mega)
```

```python
# Using regular expressions
import re
# ...................^ means starting with char not in middle or anywhere
mega = mega.loc[data['Name'].str.contains('Char', flags=re.I, regex=True)]
print(mega)
```

## Conditional Statements

```python
#Changing Type 1 with 'dragon' to 'flamethrower'
# .loc[rowindexer, columnindexer] = value
mega.loc[mega['Type 2']=='Dragon', 'Type 2'] = 'Flame Thrower'
print(mega)
```
